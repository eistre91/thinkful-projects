```python
def tf_sent_to_vec(sent_list):
    success=True
    vecs = []
    for word in sent_list:
        if word not in pres_vec.wv:
            vecs.append(np.zeros(100))
        else:
            vecs.append(pres_vec.wv[word])
    if len(vecs) == 0:
        success=False
        return None
    vec_final = vecs[0]
    for vec in vecs[1:]:
        vec_final = vec_final + vec
    return vec_final/len(vecs)
```

```python
pres_data['text_vec'] = pres_data[0].apply(tf_sent_to_vec)
```

```python
test = pd.DataFrame(pres_data_drop['text_vec'].values.tolist())
test['pres'] = pres_data_drop[1].values
```

```python
X = test.drop(['pres'], axis=1)
Y = test['pres']

X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    Y,
                                                    test_size=0.25,
                                                    random_state=42)
```

```python
lr = LogisticRegression()
train = lr.fit(X_train, y_train)
print(X_train.shape, y_train.shape)
print('Training set score:', lr.score(X_train, y_train))
print('\nTest set score:', lr.score(X_test, y_test))
```
